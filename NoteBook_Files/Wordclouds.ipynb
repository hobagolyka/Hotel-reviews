```python
import numpy as np
import pandas as pd
from wordcloud import WordCloud

import matplotlib.pyplot as plt

# load cleaned data
reviews = pd.read_csv('../Data/simpler_clean_data.csv')

reviews.head()
```

```python
print(reviews.Review[12])
```

```python
text = " ".join(str(review) for review in reviews.Review)
wordcloud = WordCloud(height=800, width=1200, background_color="white").generate(text)

plt.figure(figsize=[12,8])
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

fig_path = '../export/'
plt.savefig(fig_path + 'all_reviews_wordcloud', fmt = 'png', dpi = 100)

plt.show()
```

```python
#positive reviews
text = " ".join(str(reviews.Review[index]) for index in range(len(reviews.Review)) if reviews.Reviewer_Score[index] == 1)
wordcloud = WordCloud(height=800, width=1200,background_color="white").generate(text)

plt.figure(figsize=[12,8])
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

fig_path = '../export/'
plt.savefig(fig_path + 'pos_reviews_wordcloud', fmt = 'png', dpi = 100)

plt.show()
```

```python
#negative reviews
text = " ".join(str(reviews.Review[index]) for index in range(len(reviews.Review)) if reviews.Reviewer_Score[index] == 0)
wordcloud = WordCloud(height=800, width=1200,background_color="white").generate(text)

plt.figure(figsize=[12,8])
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

fig_path = '../export/'
plt.savefig(fig_path + 'neg_reviews_wordcloud', fmt = 'png', dpi = 100)

plt.show()
```
